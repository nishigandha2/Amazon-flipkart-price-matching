##<font color=orange>**Missing value treatment**
<font color=green>we will replace missing  brand name with unknown for both amazon and flipkart

> Indented block



```python
amazon['brand'].fillna('unknown',inplace=True)
flipkart['brand'].fillna('unknown',inplace=True)

```

```python
flipkart.isnull().sum() 
```



<font color=Green>**we are replacing retail price and discounted price by median**

```python
amazon["retail_price"].fillna(amazon["retail_price"].median(),inplace=True)
amazon["discounted_price"].fillna(amazon["discounted_price"].median(),inplace=True)
```

```python
flipkart["retail_price"].fillna(flipkart["retail_price"].median(),inplace=True)
flipkart["discounted_price"].fillna(flipkart["discounted_price"].median(),inplace=True)
```

```python
flipkart.isnull().sum() 
```

##<font color=orange> **Droping columns**
<font color=green>**we will drop unnesseary column ,using only column required for output**

```python
flipkart.drop(['crawl_timestamp', 
              'product_url', 
              'product_category_tree', 
              'pid',
              'image','brand','description',
              'is_FK_Advantage_product',
              'overall_rating','product_rating',
              'product_specifications'], axis=1, inplace=True)

amazon.drop(['crawl_timestamp', 
              'product_url', 
              'product_category_tree', 
              'description',
              'image','brand','pid',
              'is_FK_Advantage_product',
              'overall_rating','product_rating',
              'product_specifications'
             ], axis=1, inplace=True)
```

```python
flipkart.isnull().sum() 
```

```python
amazon.rename(columns={'product_name':"product name in amazon","retail_price":"Retail Price in amazon","discounted_price":"Discounted Price in amazon"},inplace=True)
amazon
```

```python
flipkart.rename(columns={'product_name':"product name in flipkart","retail_price":"Retail Price in flipkart","discounted_price":"Discounted Price in flipkart"},inplace=True)
flipkart
```

##<font color=green>**Merging matching data of amazon and flipkart**

```python
flipkart.columns = ["uniq_id",'product name in flipkart','Retail Price in flipkart','Discounted Price in flipkart']

amazon.columns = ["uniq_id",'product name in flipkart','Retail Price in amazon','Discounted Price in amazon']

df = pd.merge(flipkart, amazon, how='outer')
df = df.drop_duplicates()
df
```

```python
df.drop(['uniq_id'], axis=1, inplace=True)
```
